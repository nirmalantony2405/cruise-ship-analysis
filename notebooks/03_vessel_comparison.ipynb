```python
import pandas as pd
import matplotlib.pyplot as plt
from src.data.loader import load_data
from src.features.engineering import engineer_features

# Load cleaned data
df = load_data('../data/processed/cleaned_data.csv')

# Engineer features
df = engineer_features(df)

# Compare total propulsion power
plt.figure(figsize=(10, 6))
for vessel in df['Vessel Name'].unique():
    vessel_df = df[df['Vessel Name'] == vessel]
    plt.hist(vessel_df['Propulsion Power (MW)'], bins=50, alpha=0.5, label=vessel)
plt.title('Propulsion Power Distribution')
plt.xlabel('Propulsion Power (MW)')
plt.ylabel('Frequency')
plt.legend()
plt.savefig('../reports/figures/propulsion_power_distribution.png')
plt.show()

# Compare total electrical power consumption
plt.figure(figsize=(10, 6))
for vessel in df['Vessel Name'].unique():
    vessel_df = df[df['Vessel Name'] == vessel]
    plt.hist(vessel_df['Total Electrical Power Consumption (MW)'], bins=50, alpha=0.5, label=vessel)
plt.title('Electrical Power Consumption Distribution')
plt.xlabel('Power Consumption (MW)')
plt.ylabel('Frequency')
plt.legend()
plt.savefig('../reports/figures/electrical_power_distribution.png')
plt.show()
```
